```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

datasets.load_iris()
X=iris.data[:, :2]
y=iris.target

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)

clf=LogisticRegression(random_state=1)
clf.fit(X_train,y_train)

clf2=SVC(random_state=1)
clf2.fit(X_train,y_train)

prediction=clf.predict(X_test)
prediction2=clf2.predict(X_test)

accuracy_score(y_test,prediction),accuracy_score(y_test,prediction2)
svc_scores = cross_val_score(clf2, X_train, y_train, cv=4)
svc_scores
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.model_selection import cross_val_score

iris=datasets.load_iris()
X=iris.data
y=iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y,
stratify=y, random_state = 0)

all_scores=[]

for n_neighbors in range(1,6):
    knn_clf=KNeighborsClassifier(n_neighbors)
    all_scores.append((n_neighbors,cross_val_score(knn_clf,X_train,y_train, cv=5).mean(),cross_val_score(knn_clf,X_train,y_train,cv=5).std()))
sorted(all_scores,key= lambda x:x[1], reverse=True)
```

```python
import pandas as pd

url="https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/master/diabetes.csv"
pima = pd.read_csv(url)

column_names=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome']
feature_names=column_names[:-1]
pima.head()
#insuline/skin_thickness - 0 - something wrong
```

```python
pima.describe()
```

```python
#The features plasma_con, blood_pressure, skin_mm, insulin, bmi
#have 0s as values. These values could be physically impossible.
%matplotlib inline
pima.Pregnancies.hist(bins=50)
```

```python
pima.hist(figsize=(15,9),bins=50)
```

```python
pima.Outcome.value_counts()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X=pima[feature_names]
y=pima['Outcome']

X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=7,stratify=y)
lr=LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)

y_pred=lr.predict(X_test)

accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred,labels=[1,0])
```

```python
y_pred_proba=lr.predict_proba(X_test)
y_pred_proba
```
